```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Exercise 1 ###
# Sheriff has 180 km/h car
# Bank robber has 150 km/h car and five-minute head start
#
# How long does it take the sheriff to catch the robber?
# What distance they have traveled at that point?
#
# (ignore acceleration, traffic, etc)
####################

# Assign and convert to minutes
sheriff_speed = 180 / 60
robber_speed = 150 / 60

# X values from 0 to 50
x = np.linspace(0, 40, 1000)

# Equation
y_sheriff = sheriff_speed * (x - 5)
y_robber = robber_speed * x

plt.plot(x, y_sheriff, label='Sheriff: y = 3(x-5)')
plt.plot(x, y_robber, label='Robber: y = 2.5x')

# Add labels and title
plt.xlabel('time (minutes)')
plt.ylabel('distance (in km)')
plt.title('Line with slope 3 passing through origin')
plt.legend()

# Show the crossing point
plt.axvline(x=30, color="purple", linestyle="--")
plt.axhline(y=75, color="purple", linestyle="--")

# Show the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)

# Define X and Y range limits to display
plt.xlim(0, 40)
plt.ylim(0, 100)

plt.show()
```

```python
### Exercise 2 ###
# Jil designs solar panels as a hobby.
# On April 1st, Jills "Mark I" design begins generating power: 1 kJ/day.
# On May 1st, Jills "Mark II" design begins generating power: 4 jK/day.
#
# What day is it when Jill's Mark II design has generated as much total energy as the Mark I design?
# How much total energy have both generated by that day?
# What would the solutions to (1.) and (2.) be if Mark II design generated 1kJ of power per day?
####################

# In kJ/day
design_1 = 1
design_2 = 4

# Linear spaced days, in two months, with 1000 points
x = np.linspace(0, 60, 60)

y1 = design_1 * x
y2 = design_2 * (x - 30)

plt.plot(y1, label="Design I")
plt.plot(y2, label="Design II")

plt.xlabel('time (days)')
plt.ylabel('energy (in kj)')
plt.title('Line with slope 3 passing through origin')
plt.legend()

# Show the crossing point
plt.axvline(x=15, color="purple", linestyle="--")
plt.axhline(y=25, color="purple", linestyle="--")

# Define X and Y range limits to display
plt.xlim(0, 60)
plt.ylim(0, 100)

plt.show()
```

```python
#1. 40 days from April 1st.
#2. they generated approx 80kJ.
#3. They would have no solution (they never cross).
```

## Tensor operations

```python
import numpy as np

# Calculate transpose
x = [[42, 4, 7, 99],
     [-99, -3, 17, 22]]

x = np.array(x)

print(x.T)
```

```python
# Hadamard product of matrices
x = np.array(
    [[25, 10],
     [-2, 1]])
y = np.array(
    [[-1, 7],
     [10, 8]])

print(x*y)
```

```python
# Dot product of tensors
w = np.array([-1, 2, -2])
x = np.array([5, 10, 0])

print(np.dot(w, x))

# Or, manually, sum reductively the product of the tensors
np.sum(w*x)
```

#### Check if a matrix is orthogonal


```python
X = np.array([
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
])

# Calculate the matrix transpose
Xt = X.T

# Multiply matrix with its transpose
multiplied = X * Xt

# If the result of multiplying the matrix with its transpose is the identity matrix,
# is orthogonal
print(f"Is orthogonal: {multiplied == np.identity(X.shape[0])})")
```
